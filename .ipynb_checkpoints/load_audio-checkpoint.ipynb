```python
# plot segment of audio file based on time location index
def plot_seg(file, start, end):
    segment = file.loc[(file.index >= start) & (file.index <= end)]
    ax = segment.plot()
    return ax

# "this was easy for us"
# sent1 = plot_seg(a_df, 18000, 43000)
```
